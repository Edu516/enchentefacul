```python
import pandas as pd
import numpy as np
```

```python
# Por achar mais facil o trabalho com arquivos .csv, primeiro irei transformar os arquivos para csv
# Nomes dos arquivos
arquivos = ['ituporangaChuva', 'ituporangaNivel', 'riodosulNivel', 'taioChuva', 'taioNivel']

for arq in arquivos:
    # Ler o arquivo de texto
    with open(f'{arq}.txt', 'r') as arquivo:
        linhas = arquivo.readlines()

    # Criar listas para armazenar os dados
    data_lista = []
    hora_lista = []
    chuva_lista = []

    # Analisar as linhas do arquivo e extrair os dados
    for linha in linhas:
        partes = linha.strip().split()
        
        # Verificar o número de partes
        if len(partes) == 3:
            data, hora, chuva = partes
        elif len(partes) == 2:
            # Se houver apenas 2 partes, preencha a terceira com um valor vazio (vazio ou outra coisa que você preferir)
            data, hora = partes
            chuva = ''  # Preencha com um valor vazio
        else:
            # Trate outras situações conforme necessário
            data = hora = chuva = ''  # Preencha todas as partes com vazios ou trate de acordo com sua necessidade
        
        data_lista.append(data)
        hora_lista.append(hora)
        chuva_lista.append(chuva)

    # Criar um DataFrame do Pandas
    tipo_coluna = 'Chuva (mm)'
    if 'Nivel' in arq:
        tipo_coluna = 'Nível (cm)'
    dados = pd.DataFrame({'Data': data_lista, 'Hora': hora_lista, tipo_coluna: chuva_lista})

    # Salvar o DataFrame em um arquivo CSV
    dados.to_csv(f'{arq}.csv', index=False)
```

```python
itu_nivel = pd.read_csv('ituporangaNivel.csv');
itu_chuva = pd.read_csv('ituporangaChuva.csv');

taio_nivel = pd.read_csv('taioNivel.csv');
taio_chuva = pd.read_csv('taioChuva.csv');

riodosul_nivel =  pd.read_csv('riodosulNivel.csv');
```
